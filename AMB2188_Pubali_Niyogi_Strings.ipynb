```python
# strip- to remove spaces from the begining and end of a string
greet=" Hello Morning "
s=greet.strip()

print(greet)
print(len(greet))
print(s)
print(len(s))
```

```python
# to remove space from only one side
greet=" Hello Morning "
s1= greet.lstrip()  # .lstrip removes space from left end
s2= greet.rstrip()  # .rstrip removes space from right end

print(greet)
print(len(greet))

print(s1)
print(len(s1))

print(s2)
print(len(s2))
```

```python
# assign a variable with "Today is 19-Jan-2026 MONDAY"
#calculate the no. of vowels, consonants,blankspaces
#and other characters

s="Today is 19-Jan-2026 MONDAY"
bc=0
vc=0
cc=0
for b in s:
  if b==' ':
    bc=bc+1
  elif b=='a' or b=='e' or b=='i' or b=='o' or b=='u' or b=='A' or b=='E' or b=='I' or b=='O' or b=='U':
    vc=vc+1
  else:
    cc=cc+1

print(bc)
print(vc)
print(cc)
```
