```python
import pandas as pd
import numpy as np

#Create data with missing values
data=pd.DataFrame({
    'plot':['A','B','C','D','E'],
    'yield_kg':[2500,np.nan,2800,np.nan,2600],
    'rainfall':[150,200,np.nan,180,170],
    'temp.':[28,30,29,np.nan,27]
})

print("===Data with misssing values===")
print(data)

#Detecting missing data
print("\nDetecting missing values")
print("Is null?")
print(data.isnull())
print(f"\nTotal missing per column:")
print(data.isnull().sum())
```

```python
#Filling missing values
print("Filling strategies")

#Fill with mean
data_mean=data.copy()
print(data_mean['yield_kg'])
data_mean['yield_kg']=data_mean['yield_kg'].fillna(data_mean['yield_kg'].mean())
print("\n","Fill with mean:")
print(data_mean['yield_kg'])
```

```python
#Fill with median
data_median=data.copy()
print(data_median['rainfall'])
data_median['rainfall']=data_median['rainfall'].fillna(data_median['rainfall'].median())
print(data_median['rainfall'])
```

```python
#Forward fill(use previous value)

data_ffill=data.copy()
print(data.ffill)
data_ffill=data_ffill.fillna(method='ffill')
print("\nForward fill:")
print(data_ffill)  #ffill---> replaces missing values with the last valid (non-mising) value above it in the column
```

```python
# Dropping missing data
print("Dropping Rows")
data_dropped=data.dropna()   #it will drop the rows containing null values and print the rows without null values
print(data_dropped)
print(f"original rows: {len(data)},After drop: {len(data_dropped)}")
```
