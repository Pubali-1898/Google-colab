```python
#list is mutable
l1=[1,2,3]
print(id(l1))

l1.append(4)
print(id(l1))
```

```python
# set is mutable
s1={1,2,3}
print(id(s1))

s1.add(4)
print(s1)
print(id(s1))
```

```python
# dictionary is mutable
d1={'A':1,'B':2,'C':3,'D':4}
print(id(d1))

d1['E']= 5
print(id(d1))
```

```python
# string datatye is immutable
a="GKVK"
print(id(a))

a=a+"College"
print(id(a))
```

```python
a="Python"
print(id(a))

a=a+ 3
print(a)
print(id(a))
```

```python
a="Python"
print(id(a))

a=a+ "3"
print(a)
print(id(a))

a="Python"
a[0]= "N"
```
