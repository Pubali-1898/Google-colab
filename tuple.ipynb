```python
#Tuple declaration
T=("A","B",9,10)
print("Elements of tuple are",T)
print("Length of tuple is",len(T))
print("Type of data structure is",type(T))
```

```python
#converting list to tuple
T=tuple(["A","B",9,10])
print("Elements of tuple are",T)
print("Length of tuple is",len(T))
print("Type of data structure is",type(T))
```

```python
T=tuple(("A","B",9,10))
print("Elements of tuple are",T)
print("Length of tuple is",len(T))
print("Type of data structure is",type(T))
```

```python
#Tuple with one item
T=("Pubali",)      # if there is a single word in a tuple it has to be ended with a comma
print("Elements of tuple are",T)
print("Length of tuple is",len(T))
print("Type of data structure is",type(T))
```

```python
T=("Pubali")    # if comma isn't given it will be considered as a string datatype
print("Elements of tuple are",T)
print("Length of tuple is",len(T))
print("Type of data structure is",type(T))
```

```python
#Tuple iteration using while loop
T=("A","B",9,10)
print("T-->",T)

i=0
print("")
while i < len(T):
  print(T[i])
  i+=1
```

```python
#Tuple iteration using for loop
T=("A","B",9,10)
print("T-->",T)

print("")
for i in range (len(T)):  #refering to the indexes
  print(T[i])
```

```python
T=("A","B",9,10)
print("T-->",T)

print("")
for i in range (len(T)):
  print(i)
```

```python
T=("A","B",9,10)
print("T-->",T)

print("")
for s in T:  #refering to the elements
  print(s)
```

```python
# to access individual element in tuple
T=(3,6,9,12,5,6,1)
print("T-->",T)

print("T[1]-->",T[1])
print("T[2]",T[2])
print("T[-1]",T[-1])
print("T[-7]",T[-7])
print("T[7]",T[7])
```

```python
T=(3,6,9,12,5,6,1)
print("T-->",T)

print("T[1:3]-->",T[1:3])
print("T[2:]",T[2:])
print("T[:2]",T[:2])
print("T[2:5]-->",T[2:5])
print("T[-4:-1]",T[-4:-1])
print("T[:-1]",T[:-1])
print(T[-1:-4])
```

```python
T=(3,6,9,12,5,10,1)
print("T-->",T)

print("T[::1]",T[::1])
print("T[::-1]",T[::-1])
print("T[::-2]",T[::-2])
print("T[::2]",T[::2])
```
