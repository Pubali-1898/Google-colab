```python
# create a new list with each element increased by 2
numbers=[10,20,30,40,50,60]
for i in numbers:
  print(i+2)
print(numbers)
```

```python
numbers=[10,20,30,40,50,60]
n=[] #empty list
for i in numbers:
  n.append(i+2)
print("New list",n)
print("Original list",numbers)
```

```python
#list comprehension- create a new list with each element increased by 2
numbers=[10,20,30,40,50,60]
n=[i+2 for i in numbers] #n= expression of numbers in iterable
print("New list",n)
print("original list",numbers)
```

```python
#Create a new list containing no.s completely divisible by 4
numbers=[10,20,30,40,50,60]
n=[]
for i in numbers:
  if i%4==0:
    n.append(i)
print("numbers divided by 4",n)
print("original list",numbers)
```

```python
#list comprehension of above program
numbers=[10,20,30,40,50,60]
n=[i for i in numbers if i%4==0 ]
print("Numbers divided by 4 ",n)
print("Original list",numbers)
```

```python
numbers=[10,20,30,40,50,60]
n=[i for i in numbers if i%4==22]
print("Numbers divided by 4 ",n)
print("Original list",numbers)
```

```python
#print even or odd using list comprehension and if else
numbers=[10,20,30,41,51,61]
n=["even"if i%2==0 else"odd" for i in numbers]#n=[expression1 if condition else Expression2 for item in iterable]
print(n)
```
