```python
# assign a variable with "Today is 19-Jan-2026 MONDAY"
#calculate the no. of vowels, consonants,blankspaces
#and other characters

s="Today is 19-Jan-2026 MONDAY"
bc=vc=cc=d=oc=0
for b in s:
  if b.isspace():
    bc=bc+1
  elif b in 'aeiouAEIOU' :
    vc=vc+1
  elif b.isalpha():
    cc=cc+1
  elif b.isdigit():
    d=d+1
  else:
    oc=oc+1

print("spaces:",bc)
print("vowels:",vc)
print("consonants:",cc)
print("digits:",d)
print("other characters:",oc)
```
