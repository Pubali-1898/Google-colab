```python
# Most effective pesticide
psc=["Neem oil","Chlorpyrifos","Malathion","Imidacloprid","Spinosad"]
e=[75,85,70,90,88]
print("Most effective pesticide is: ")

mx= max(zip(e,psc))
print(mx[1])
```

```python
#Least effective pesticide
print("Least effective pesticide is: ")
mn=min(zip(e,psc))
print(mn[1])
```

```python
#Average of effectiveness
avg=sum(e)/len(e)
print("Average efectiveness of pesticides= ",avg,"%")
```

```python
#effectiveness equal to or above average
for i,j in zip(e,psc):
  if i>=avg:
    print(j,"-",i,"%")
```

```python
#list all pesticides with effectivenes >80%
print("Pesticides with effectiveness >80% are: ")
for i,j in zip(e,psc):
  if i>80:
    print(j,"-",i,"%")
```

```python
#Count the no. of pesticides with effectivess<80%
psc=["Neem oil","Chlorpyrifos","Malathion","Imidacloprid","Spinosad"]
e=[75,85,70,90,88]
for i,j in zip(e,psc):
  if i<80:
    l=e.count(i)
    print(j,l)
```

```python
#pesticides suitable for organic farming
print("pesticides suitable for organic farming are:")
for i,j in zip(e,psc):
  if i>=85:
    print(j)
```

```python
#Increase effectiveness of all pesticides by 5%
for i,j in zip(e,psc):
  print(j,"-",i+5,"%")
```

```python
#replace pesticides with effectiveness <75% as "non-recommendable"
for i,j in zip(e,psc):
  if i<75:
    print(j,i,"%""-Non-recommendable")
```

```python
 # Sort the pesticides based on their effectiveness in ascending order
psc=["Neem oil","Chlorpyrifos","Malathion","Imidacloprid","Spinosad"]
e=[75,85,70,90,88]
print("Effectiveness of pesticides arranged in ascending order:")
s=sorted(zip(e,psc))
print(s)
```
