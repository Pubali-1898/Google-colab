```python
#Increase in cost of farm equipment by 8%
#using .append
eqp=["Tractor","Plough","Harvester","Seeder","Sprayer","Thresher"]
c=[850000,25000,1200000,60000,18000,95000]
new_cost=[]
for i in c:
  new_cost.append(i*1.08)
print("Actual cost= ",c)
print("New cost= ",new_cost)
```

```python
#using append and rounding off the decimal value
eqp=["Tractor","Plough","Harvester","Seeder","Sprayer","Thresher"]
c=[850000,25000,1200000,60000,18000,95000]
new_cost=[]
for i in c:
  new_cost.append(round((i*1.08),2))
print("Actual cost= ",c)
print("New cost= ",new_cost)
```

```python
#same program as above using list comprehension
eqp=["Tractor","Plough","Harvester","Seeder","Sprayer","Thresher"]
c=[850000,25000,1200000,60000,18000,95000]
new_cost=[int(i*1.08)for i in c]
print("Updated cost",new_cost)
```

```python
#replace equiment with cost >10,00,000 as "High Investment"
eqp=["Tractor","Plough","Harvester","Seeder","Sprayer","Thresher"]
c=[850000,25000,1200000,60000,18000,95000]
New_equipment=[]
for i in range (len(eqp)):
  if c[i]>1000000:
    New_equipment.append("High Investment")
  else:
    New_equipment.append(eqp[i]) # if I had typed it as New_equipment.append(i), in output I will get only index values of equipments
print("The updated list :",New_equipment)
```

```python
# doing above program using list comprehension
New_equipment=["High Investment"if i>1000000 else j for i,j in zip(c,eqp)]
print("The updated list : ",New_equipment)
```

```python
# Create cost category list
category=[] #taking an empty list to store new components
for i in c:
  if i<100000:
    category.append("Low")
  elif i<=500000:
    category.append("Medium")
  else:
    category.append("High")
print("Equipments are: ",eqp)
print("Cost of equipments: ",c)
print("Cost categories",category)
```

```python
#doing same program using list comprehension
category=["Low" if i<100000 elif i<=500000 "Medium" else "High" for i in c]
print("Equipments are: ",eqp)
print("Cost of equipments: ",c)
print("Cost categories",category)
```

```python
# Reduce the cost of equipment by 15% for subsidy eligible equiment
# Assuming subsidy for equipment costing below Rs100000
subsidy_cost=[]
for i in c:
  if i<100000:
    subsidy_cost.append(int(i*0.85))
  else:
    subsidy_cost.append(i)
print("New cost after providing subsidy",subsidy_cost)
```

```python
# doing above program using list comprehension
subsidy_cost=[int(i*0.85) if i<100000 else i for i in c]
print("New cost after providing subsidy",subsidy_cost)
```
