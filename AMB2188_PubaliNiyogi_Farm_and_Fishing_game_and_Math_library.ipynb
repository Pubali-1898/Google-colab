```python
#Farm game
print("Welcome to Farm Game!!")
print("1.Plants sown 🌿")
print("2.Water crops💦")
print("3.Harvest crops🌻")
print("4.Exit")
crops=0
points=0

while(True):
  ch=int(input("Enter your choice"))
  if ch==1:
    crops+=1
    if crops>=1:
      print("Plants sown😇",crops)
    else:
      print("Not a single seed sown")
  elif ch==2:
    if crops>=1:
      print("Irrigation done!💧 ")
    else:
      print("No crops to irrigate")
  elif ch==3:
    if crops>=1:
      points=points+(crops*10)
      crops=0
      print("Successfully Harvesting done!👍")
      print("You have earned",points,"points ✌️")
      print("Congratulations!🤠")
    else:
      print("No crop to harvest🤦‍♀️")
  elif ch==4:
    print("Exit🙅‍♀️")
    print("You have earned",points,"points ✌️")
    break
  else:
    print("Invalid choice😵‍💫")
```

```python
#Catch fish
print("Welcome to the world of fishing")
print("1.Catch fish")
print("2.Display the no. of fishes he has")
print("3.Marketting of fishes")

earnings=0
fish=0

while(True):
  d={}
  i=input("Enter your choice")
  if i==1:
    fish+=1
    if fish>=1:
      n=input("Enter the name of fish")
      d[n]= fish
      print("updated data:",d)
    else:
      print("No fish has been caught")
  elif i==2:
    if fish>=1:
      earnings=earnings+(fish*earnings)
      earnings=0
      print("Income",earnings)
    else:
      print("No income")
  elif i==3:
      print("Exit")
      print("Income",earnings)
      break
  else:
     print("Invalid choice")
```
