```python
#scope of variables
def func(x):
  x=50
  print("Local x is:",x)
  x=2
  print("Changed local x to :",x)
x=50
func(x)
print("X is still:",x)
```

```python
def func():
  y=2
  print("Changed y to:",y)
y=50
func()
print("Y is still:",y)
```

```python
#global variable
def func():
  global y
  print("Y is:",y)
  y=2
  print("Changed value to:",y)

  y=50
  func()
  print("Value of y is:",y)
```

```python
total=0
#function definition is here
def sum(arg1,arg2):
  #Adding both the parameters and returning them
  total= arg1+arg2  #total is a local variable here
  print("Inside the function total:",total)
  return total

sum(10,20) #we are calling the sum function
print("Outside the functon total :",total)
```

```python
#Required argument
def printme(y):    #printme is a function name
  print(y)

printme("Hello")
printme(1234)
```

```python
def printme(str):    #printme is a function name
  print(str)
  return
printme("Hello")
printme(1234)
#printme("Hello","India") # throws an error bcoz in function header only one variable has been given but user here is giving 2 variables
```

```python
#named argument
def printme(str):    #printme is a function name
  print(str)
  return
printme("Hello")
printme(1234)
printme(str="I am proud to be an Indian")  #named argument
#printme(abc="Hello") # will throw error bcoz abc is an unknown argument
```

```python
#Keyword argument
def hello(a,b=10,c=20): #default argument
 # we have given default values to b and c, if in case user wont give any value for b and c, default value will be printed
  print("a is:",a,"b is:",b,"c is:",c)

# hello(a,b,c)---> throws error saying 'name a is not defined'
hello(3,6,9)  #required arguments
hello(3,100)    #a,b- required c- default
# c is the default argument
hello(3,b=7,c=10)  ## b value got changed to user given value
hello(c=200,a=100) #keyword argument; when keyword is mentioned it can be writen in any order
```

```python
def h(name,age=35):
  print("Name:",name)
  print("age:",age)
  return

h("Virat",50)
h("Virat")
h(age=35)
```

```python
#variable length argument
def check (name,*marks):  # at 1st required argument should be there
  print(name)
  print(marks)
check("ABC",10,20,30)
```
